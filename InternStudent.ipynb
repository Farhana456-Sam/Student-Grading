# **Univariate Analysis**

```python
#distribution univariate for numerical columns
plt.figure(figsize=(15,10))
for i,column in enumerate(['Age', 'Year of Study', 'Attendance', 'Projects',
       'Assignments_Submitted', 'Participation_Score', 'Project_Grade',
       'Skill_Proficiency', 'Quiz Mark', '10th Mark%', '12th Mark%'],1):
  plt.subplot(3,4,i)
  sns.histplot(df[column],kde=True)
  plt.title(f'Distribution of {column}')

plt.tight_layout()
plt.show()
```

```python
col_cat.columns
```

```python
cat_columns = df[['Gender', 'Department', 'College Name', 'Exam Grade']]
```

```python
for col in cat_columns:
    plt.figure(figsize=(8, 4))
    df[col].value_counts().plot(kind="bar", color="skyblue", edgecolor="black")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])

# Compute and plot correlation matrix
plt.figure(figsize=(8, 8))
dataplot = sns.heatmap(numeric_df.corr(), annot=True)
plt.show()
```
